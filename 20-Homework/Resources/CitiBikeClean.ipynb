```python
import pandas as pd
```

```python
# The path to our CSV file
file = "201902-citibike-tripdata.csv"

# Read our Citi Bike data into pandas
df = pd.read_csv(file)
df.head()
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
# Extract "tripduration", "starttime", "stoptime", "start station latitude", "start station longitude", "end station latitude",
# "end station longitude", "bikeid", "usertype", "birth year", and "gender"
reduced_citibike_df = df.loc[:, ["tripduration", "starttime", "stoptime",
                                    "start station latitude", "start station longitude",
                                    "end station latitude", "end station longitude",
                                    "bikeid", "usertype", "birth year", "gender"]]
reduced_citibike_df
```

```python
# Remove rows without a specified gender
reduced_citibike_df = reduced_citibike_df.loc[(
    reduced_citibike_df["gender"] > 0)]
reduced_citibike_df.head()
```

```python
# Create a new column that finds the approximate age of each user
reduced_citibike_df["age"] = 2019 - reduced_citibike_df['birth year']
```

```python
reduced_citibike_df.head()
```

```python
# Export the cleaned dataframe into a new CSV file
reduced_citibike_df.to_csv(r'data.csv')
```
